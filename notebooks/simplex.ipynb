```python
from scipy.optimize import linprog

# Maximizar Z = 3x + 2y  =>  Minimizar -3x - 2y
c = [-3, -2]

# Restricciones de desigualdad (A_ub * x <= b_ub)
# 2x + y <= 10
# x + 2y <= 12
A = [[2, 1], [1, 2]]
b = [10, 12]

# Límites de las variables (x, y >= 0)
x_bounds = (0, None)
y_bounds = (0, None)

res = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

print(f"Estado: {res.message}")
print(f"Valores óptimos: x = {res.x[0]}, y = {res.x[1]}")
print(f"Valor máximo de Z: {-res.fun}")
```
